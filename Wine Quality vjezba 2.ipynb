```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,classification_report

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.svm import SVC,SVR
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
target.unique()
```

```python
linreg=LinearRegression()

linreg.fit(X_train,y_train)
```

```python
y_pred=linreg.predict(X_test)
```

```python
print( " Model Evaluation Linear R : mean absolute error is ", mean_absolute_error(y_test,y_pred))
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(X_train,y_train)
```

```python
y_predLR=logreg.predict(X_test)
```

```python
print(classification_report(y_predLR,y_test,zero_division=1))
```

```python
tree=DecisionTreeClassifier()
```

```python
tree.fit(X_train,y_train)
```

```python
y_predDT=tree.predict(X_test)
```

```python
print(classification_report(y_predDT,y_test,zero_division=1))
```

```python
from sklearn.metrics import confusion_matrix

matrix=confusion_matrix(y_predDT,y_test)
```

```python
print(matrix)
```

```python
features['quality'].unique()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax = sns.heatmap(matrix, annot=True, cmap='Blues')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nWine quality')
ax.set_ylabel('Grades ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['4','5', '6','7','8'])
ax.yaxis.set_ticklabels(['4','5', '6','7','8'])

## Display the visualization of the Confusion Matrix.
plt.show()
```
